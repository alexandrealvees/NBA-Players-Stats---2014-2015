```python
# 5 jogadores com mais pontos
df_novo.groupby('Collage')['PTS'].max().head(10).plot.barh(title='Total de Pontos por F A C U L D A D E')
```

```python
# 10 MAiores Pontuadores da Temporada
df[['Name', 'Collage', 'PTS']].nlargest(10, 'PTS')
```

```python
maiores_pontuadores = df.groupby(['Name'])['PTS'].sum() \
                                                       .reset_index(name='Pontos') \
                                                       .sort_values(['Pontos'], ascending=False)
                             
maiores_pontuadores
```

```python
import matplotlib.pyplot as plt


plt.bar(maiores_pontuadores['Name'].head(3), maiores_pontuadores['Pontos'].head(3))
plt.title('Maiores Pontuadores')
plt.xlabel('Nomes dos Jogadores')
plt.ylabel('Total de Pontos')
plt.show()
```
